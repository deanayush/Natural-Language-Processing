## Average Number of Tokens

```python
l = []
l2 = []
count1 = 0
count2 = 0

for i in range(len(data)):
    tokenized = re.compile(r'\s+').split(data['TEXT'][i])
    #tokenized=word_tokenize(data['TEXT'][i])
    #l.append(tokenized)
    
    for j in tokenized:
        if(data['LABEL'][i] == 1):
            for k in re.findall(r"\w+",j):
                l.append(k)
                count1 = len(l)
        else:
            for k in re.findall(r"\w+",j):
                l2.append(k)
                count2 = len(l2)

total_token_count = count1 + count2
print("Total number of tokens: " + str(total_token_count))
print("Total number of tokens in LABEL 1: " + str(count1))
print("Total number of tokens in LABEL 0: " + str(count2))
print("Avg number of tokens for Label 1: " + str((count1) / total_token_count))
print("Average number of tokens for Label 0: " + str((count2) / total_token_count))
```

## Total number of words starting with consonants and vowels. 

```python
vowel_1_list = []
vowel_0_list = []
const_1_list = []
const_0_list = []
for i in range(len(data)):
    tokenized = re.compile('\s+').split(data['TEXT'][i])    
    for j in tokenized:
        if re.findall(r'^[aeiouAEIOU]\w+',j):
            if(data['LABEL'][i] == 1):
                for k in re.findall(r"^[aeiouAEIOU]\w+",j):
                    vowel_1_list.append(k)
            else:
                for k in re.findall(r"^[aeiouAEIOU]\w+",j):
                    vowel_0_list.append(k)
        
        if re.findall(r'[bcdfgjklmnpqstvxzhrwyBCDFGHJKLMNPQSTVXZHRWY]\w+',j):
            if(data['LABEL'][i] == 1):
                for k in re.findall(r"[bcdfgjklmnpqstvxzhrwyBCDFGHJKLMNPQSTVXZHRWY]\w+",j):
                    const_1_list.append(k)
            else:
                for k in re.findall(r"[bcdfgjklmnpqstvxzhrwyBCDFGHJKLMNPQSTVXZHRWY]\w+",j):
                    const_0_list.append(k)



print("\n Words starting with vowels are in LABEL 1: " + str(len(vowel_1_list)))
print("\n Words starting with vowels are in LABEL 0: " + str(len(vowel_0_list)))
print("\n Words starting with const are in LABEL 1: " + str(len(const_1_list)))
print("\n Words starting with const are in LABEL 0: " + str(len(const_0_list)))
#print(const_0_list)
```

## Total number of unique tokens after lowercase the text

```python
# num of unique tokens after lowercase
lower = []
l2 = []
lower2 = []
l1 = []
# find all tokens
for i in range(len(data)):
         tokenized = re.compile(r'\s+').split(data['TEXT'][i])
         for j in tokenized:
                if(data['LABEL'][i] == 0):
                    for k in re.findall(r"\w+", j):
                        lower.append(k)
                else:
                    for k in re.findall(r"\w+", j):
                        lower2.append(k)

# lowercase the tokens
for i, val in enumerate(lower):
    res = re.sub(r"[A-Z]", lambda x : x.group(0).lower(), val)
    l2.append(res)


# take unique tokens
unique_set = set(l2)
count = 0
unique_list = (list(unique_set))
for x in unique_list:
    #print(x)
    count += 1
print("\n Total number of unique tokens after lowercase for label 0: " + str(count)) 

for i, val in enumerate(lower2):
    res = re.sub(r"[A-Z]", lambda x : x.group(0).lower(), val)
    l1.append(res)

unique_set2 = set(l1)
count2 = 0
unique_list2 = (list(unique_set2))
for x in unique_list2:
    #print(x)
    count2 += 1
print("\n Total number of unique tokens after lowercase for label 1: " + str(count2)) 
```
